```python
import ColabTurtlePlus.Turtle as cTurtle 

cTurtle.clearscreen()
cTurtle.setup(400, 400)
cTurtle.showborder()
t = cTurtle.Turtle()
t.shape('turtle')
t.pensize(2)
t.speed(13)

def half_leaf():
    t.penup()
    t.setposition(0,0)
    t.pendown()
    
    
    t.left(90)
    t.forward(50)

    t.right(100)
    t.forward(50)

    t.left(130)
    t.forward(20)

    t.right(90)
    t.forward(50)

    t.left(100)
    t.forward(15)

    t.right(80)
    t.forward(30)

    t.left(130)
    t.forward(30)

    t.right(90)
    t.forward(15)

    t.left(120)
    t.forward(40)

    t.right(120)
    t.forward(50)

    t.left(120)
    t.forward(15)

    t.right(90)
    t.forward(30)

color_list = ['red', 'orange', 'yellow', 'green', 'blue', 'violet']

for color in color_list:
    t.pencolor(color)
    half_leaf()
    t.right(60)
    
t.penup()
t.setposition(500, 500)
```
